```python
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
z = np.sin(x)
```

```python
x
y=x * 2
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
fig , axes = plt.subplots(nrows=4,ncols=4)
axes[0][0].plot(x,y,'g')
axes[0][0].set_title('green')
axes[0][0].set_xlabel('x green')
axes[0][0].set_ylabel('y green')
axes[1][0].plot(z,y,'m')

fig.suptitle('the B G ' , fontsize = 16);

fig.subplots_adjust(top=1.5,hspace=5.5,wspace=5.5)
```

```python
axes.shape
```

```python
fig=plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.3,0.3,0.1,0.1])
axes1.plot (x,y,'r',ls='-.',lw=5)
axes2.plot(x,y,color='black',ls='-',lw=5)
axes1.legend()
axes2.legend()
```
